```python
import numpy as np
import sys
import matplotlib.pyplot as plt
from matplotlib import rcParams
import pandas as pd
import seaborn as sns
```

```python
path = 'adult.csv'
dataset = pd.read_csv(path)
```

```python
dataset.head(3)
```

```python
df = dataset.replace('?',np.nan)
df['occupation'] = df['occupation'].fillna('Prof-specialty')
df['workclass'] = df['workclass'].fillna('Private')
df['native.country'] = df['native.country'].fillna('United-States')
```

```python
df.isnull().sum()
```

```python
color = plt.get_cmap('RdYlGn')   # default color
color.set_bad('lightblue')
sns.heatmap(df.corr(), annot=True, cmap=color); 
```

```python
categorical_features = ["education","marital.status","occupation","relationship","race","sex","native.country","income"]

for i, categorical_feature in enumerate(df[categorical_features]):
    df.groupby(categorical_feature).size().plot(kind='bar')
    plt.show()
```

```python
numeric = [ 'fnlwgt','capital.gain', 'capital.loss']
#'age', 'education.num', 'hours.per.week'
for i in range(len(numeric)):
    sns.distplot(df[numeric[i]])
    if i<3:
        plt.xscale('log')
        plt.yscale('log')
    plt.show()
```

```python
# plt.rcParams["figure.figsize"] = (15,8)
for i in range(len(numeric)):
    df.boxplot(column=numeric[i])
    if i<3:
        plt.xscale('log')
        plt.yscale('log')
    plt.show()
```
